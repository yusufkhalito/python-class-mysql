```python
# collaborative_project.py

import mysql.connector

class DatabaseConnection:
    def __init__(self, host, user, port, password, database):
        self.host = host
        self.user = user
        self.port = port
        self.password = password
        self.database = database
        self.connection = None

    def connect(self):
        self.connection = mysql.connector.connect(
            host=self.host,
            user=self.user,
            port = self.port,
            password=self.password,
            database=self.database
        )

    def execute_query(self, query, data=None):
        cursor = self.connection.cursor()
        if data:
            cursor.execute(query, data)
        else:
            cursor.execute(query)
        result = cursor.fetchall()
        cursor.close()
        return result

    def commit_and_close(self):
        self.connection.commit()

        
class Book(DatabaseConnection):
    def __init__(self, title, author, isbn, db_connection):
        super().__init__(db_connection.host, db_connection.user, db_connection.port, db_connection.password, db_connection.database)
        self.title = title
        self.author = author
        self.isbn = isbn

    def add_book(self):
        query = "INSERT INTO books (title, author, isbn) VALUES (%s, %s, %s)"
        data = (self.title, self.author, self.isbn)
        self.connect()
        self.execute_query(query, data)
        self.commit_and_close()

    def search_book(self, title):
        query = "SELECT * FROM books WHERE title = %s"
        data = (title,)
        self.connect()
        result = self.execute_query(query, data)
        self.close_connection()
        return result

    def delete_book(self, book_id):
        query = "DELETE FROM books WHERE id = %s"
        data = (book_id,)
        self.connect()
        self.execute_query(query, data)
        self.commit_and_close()

class Member(DatabaseConnection):
    def __init__(self, name, address, membership_status, db_connection):
        super().__init__(db_connection.host, db_connection.user, db_connection.port, db_connection.password, db_connection.database)
        self.name = name
        self.address = address
        self.membership_status = membership_status

    def add_member(self):
        query = "INSERT INTO members (name, address, membership_status) VALUES (%s, %s, %s)"
        data = (self.name, self.address, self.membership_status)
        self.connect()
        self.execute_query(query, data)
        self.commit_and_close()

    def search_member(self, name):
        query = "SELECT * FROM members WHERE name = %s"
        data = (name,)
        self.connect()
        result = self.execute_query(query, data)
        self.close_connection()
        return result

    def delete_member(self, member_id):
        query = "DELETE FROM members WHERE id = %s"
        data = (member_id,)
        self.connect()
        self.execute_query(query, data)
        self.commit_and_close()

# Example Usage
project = DatabaseConnection("local host", "root", 3306,"(Omattaiiii1111$)" "libary system")

# Connect to the database
project.connect()

try:
    while True:
        print("\n1. Add Book\n2. Search Book\n3. Delete Book\n4. Add Member\n5. Search Member\n6. Delete Member\n7. Exit")
        choice = input("Enter your choice (1-7): ")

        if choice == '1':
            title = input("Enter book title: ")
            author = input("Enter book author: ")
            isbn = input("Enter book ISBN: ")

            book_instance = Book(title, author, isbn, project)
            book_instance.add_book()
            print("Book added successfully!")

        elif choice == '2':
            title_to_search = input("Enter the title of the book to search: ")
            book_instance = Book(None, None, None, project)
            result = book_instance.search_book(title_to_search)
            print("Search result:")
            for row in result:
                print(row)

        elif choice == '3':
            book_id_to_delete = input("Enter the ID of the book to delete: ")
            book_instance = Book(None, None, None, project)
            book_instance.delete_book(book_id_to_delete)
            print("Book deleted successfully!")

        elif choice == '4':
            name = input("Enter member name: ")
            address = input("Enter member address: ")
            status = input("Enter membership status: ")

            member_instance = Member(name, address, status, project)
            member_instance.add_member()
            print("Member added successfully!")

        elif choice == '5':
            name_to_search = input("Enter the name of the member to search: ")
            member_instance = Member(None, None, None, project)
            result = member_instance.search_member(name_to_search)
            print("Search result:")
            for row in result:
                print(row)

        elif choice == '6':
            member_id_to_delete = input("Enter the ID of the member to delete: ")
            member_instance = Member(None, None, None, project)
            member_instance.delete_member(member_id_to_delete)
            print("Member deleted successfully!")
except exception as e:
    print(f"an error occured:{e}")
            
finally:

except exception as e:
    print(f" an error occured)
           
# Close the database connection
project.close_connection()
5
```
